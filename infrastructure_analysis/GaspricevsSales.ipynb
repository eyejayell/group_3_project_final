```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress

# File to Load
Gasprice_data = "gas_prices_cleaned.csv"
Carsales_data = "Ev_sales_cleaned.csv"

# Read Data
Gasprice = pd.read_csv (Gasprice_data)
Carsales = pd.read_csv(Carsales_data)

# Combine the data into a single dataset.
Gasprice_Carsales = pd.merge (Carsales,Gasprice, how = "left", on=['Cleaned Up Date'] )
Gasprice_Carsales.dtypes 



```

```python
Gasprice_Carsales.info()
```

```python
Gasprice_Carsales.dropna(inplace=True)
```

```python
# Scatterplot of gasprice and BEV sold
x_values = Gasprice_Carsales["BEV"]
y_values = Gasprice_Carsales[ "Weekly U.S. All Grades All Formulations Retail Gasoline Prices  (Dollars per Gallon)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope, 5)) + "x + " + str(round(intercept, 5))
plt.scatter(x_values, y_values)
plt.plot(x_values, regress_values, "r-")
plt.annotate(line_eq, (2.0, 80000), fontsize=12, color="red")
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
